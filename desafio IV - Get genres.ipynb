```python
import pandas as pd
import numpy as np
import os
```

```python

classical = 'C:\DH\ds_blend_students_2020\desafio IV\MIDI folders\midi_selection\ClassicRock'

# the dictionary to pass to pandas dataframe
dict = {}
count_files = 0
      
for root, dirs, files in os.walk(classical, topdown=False):
    for name_file in files:
        dict[count_files] = {"file": root, "file_name": name_file, "compositor": 'All'}
        count_files = count_files + 1

df_files = pd.DataFrame.from_dict(dict, "index")
display(df_files)
```

```python
df_files.file.unique()
```

```python
bands = []
for row in df_files.file.unique():
    bands.append(row.split('\\')[7])
bands
```

```python
import wikipedia
```

```python
bands
```

```python
from contextlib import suppress
dict_bands = {}
for band in bands:
    with suppress(Exception):
        wiki = wikipedia.page(wikipedia.search(band + " artist")[0])
        dict_bands[band] = wiki.content[0:400]
        print(band)

```

```python
df_dict_bands = pd.DataFrame.from_dict(dict_bands, "index")
df_dict_bands
```

```python
df_dict_bands.to_csv('bands_wiki.csv')
```
